```python
from pytube import YouTube

#download a video 
yt = YouTube('https://www.youtube.com/watch?v=AutW87v5_6I&t=320s').streams.first().download(filename='mp4/testyt')
```

```python
import whisper

model = whisper.load_model("base")
result = model.transcribe("./mp4/testyt.mp4")
print(result["text"])
```

```python
import json

# write out the text from the processed mp4 file
with open('./json/testyt.json', 'a') as fp:
    fp.write(json.dumps(result))
```
